```python
from PIL import Image, ImageDraw, ImageFont
import os


os.makedirs('training_data', exist_ok=True)


texts = [
    "The quick brown fox jumps over the lazy dog",
    "Pack my box with five dozen liquor jugs",
    "Jackdaws love my big sphinx of quartz",
    "The five boxing wizards jump quickly"
]

Fonts = []
font_dir = '/Users/adityachavan/Desktop/Acc/fonts-main/apache/'  


font_name = 'Montez-Regular.ttf'  


for i, text in enumerate(texts):
    font_path = font_dir    
    try:
        font = ImageFont.truetype(font_path, 40)
        image = Image.new('RGB', (800, 200), color = (255, 255, 255))
        draw = ImageDraw.Draw(image)
        draw.text((10, 80), text, font=font, fill=(0, 0, 0))
        
        
        image_path = f'training_data/{font_name}_{i}.png'
        image.save(image_path)

        
        with open(f'training_data/{font_name}_{i}.txt', 'w') as f:
            f.write(text)
    except Exception as e:
        print(f"Error with font {font_name}: {e}")
```

```python
# This script should be run in the terminal

# Create training data
for img in training_data/*.png; do
    base=$(basename $img .png)
    tesseract $img $base nobatch box.train
done

# Compute the character set
unicharset_extractor training_data/*.box

# Create font properties file
echo "Roboto-Regular 0 0 0 0 0" > font_properties

# Clustering
shapeclustering -F font_properties -U unicharset *.tr

# MFTraining
mftraining -F font_properties -U unicharset -O eng.unicharset *.tr

# CNTraining
cntraining *.tr

# Move the generated files
mv inttemp eng.inttemp
mv normproto eng.normproto
mv pffmtable eng.pffmtable
mv shapetable eng.shapetable

# Combine the trained data files into a traineddata file
combine_tessdata eng.
```
