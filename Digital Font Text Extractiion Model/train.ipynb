```python
from PIL import Image, ImageDraw, ImageFont
import os


font_dir = '/Users/adityachavan/Desktop/Acc/fonts-main/apache/'


os.makedirs('training_data', exist_ok=True)




def generate_images_for_font(font_path, font_name):
    for i, text in enumerate(texts):
        try:
            font = ImageFont.truetype(font_path, 40)
            image = Image.new('RGB', (800, 200), color=(255, 255, 255))
            draw = ImageDraw.Draw(image)
            draw.text((10, 80), text, font=font, fill=(0, 0, 0))
            
            
            image_name = f'{font_name}_{i}.png'
            image_path = os.path.join('training_data', image_name)
            image.save(image_path)

            
            with open(f'{image_path}.txt', 'w') as f:
                f.write(text)
        except Exception as e:
            print(f"Error with font {font_name}: {e}")


for root, dirs, files in os.walk(font_dir):
    for file in files:
        if file.endswith(".ttf") or file.endswith(".otf"):
            font_path = os.path.join(root, file)
            font_name = os.path.splitext(file)[0]
            generate_images_for_font(font_path, font_name)
```

```python
import os
import pytesseract
from PIL import Image


os.environ['TESSDATA_PREFIX'] = '/opt/homebrew/share/tessdata'


pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'


custom_oem_psm_config = r'--oem 3 --psm 3 -l eng'


image_path = 'training_data/Montez-Regular_1.png'
image = Image.open(image_path)


extracted_text = pytesseract.image_to_string(image, config=custom_oem_psm_config)
print("Extracted Text:")
print(extracted_text)
```
